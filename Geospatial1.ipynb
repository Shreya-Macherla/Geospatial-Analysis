## Task 1.2.5: Present a scatter plot between mean population of each country and mean life expectancy considering from year 2000 until 2010. Very briefly interpret the generated plot, particularly look for any correlation (if present) among the plotted variables. In this question, you should consider each year between 2000 and 2010 to find the mean population and mean life expectancy

```python
# Importing statistics library for calculating mean.
import statistics
from statistics import mean

# Extracting required coulmns(country code, 2000-2010) from life expectancy dataframe.
LEdata125= LEdata[['Country Code','2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010']]
LEdata125
```

```python
# Extracting required coulmns(country code, 2000-2010) from population dataframe.
popdata125= popdata[['Country Code','2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010']]
popdata125
```

```python
# Calculationg mean LE of all countries from 2000-2010
final_mean_le = LEdata125.mean(axis=None, skipna=None, level=None, numeric_only=None)
final_mean_le
```

```python
# Calculationg mean Population of all countries from 2000-2010
final_mean_pop = popdata125.mean(axis=None, skipna=None, level=None, numeric_only=None)
final_mean_pop
```

```python
# Visualising mean LE data for years 2000 to 2010.
final_mean_le.plot()
```

```python
# Visualising mean population data for years 2000 to 2010.
final_mean_pop.plot()
```

### Scatter plot between mean population of each country and mean life expectancy considering from year 2000 until 2010.

```python
plt.ylabel('Mean LE  2000 - 2010')
plt.xlabel('Mean population 2000 - 2010')
plt.title("Mean Population of each Country and their Mean Life expentancy considering from year 2000 until 2010")
plt.scatter(final_mean_pop, final_mean_le)
```

From the plot we can see that both the life expectancy and population has been gradually increasing from 2000 to 2010. They are proportional to each other.
This could be due to technology advancements in various secctors of life like health care etc.
